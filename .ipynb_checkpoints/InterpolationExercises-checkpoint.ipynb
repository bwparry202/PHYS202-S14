2. Thermistors are electrical devices that measure temperature. They are made out of a material whose resistance changes very predictably with temperature. In the lab we measure the resistance of the thermistor and then compute the temperature from the engineering data that the manufacturer sends with the devices. The table below is data for a set of thermistors used in PHYS 340. The temperature values are in ∘C and the resistances in mega-Ohms, MΩ.

(a) Let's turn this look-up table into a continuous function. Write a reusable function Temp(Resistance) that takes any value of Resistance and returns the interpolated temperature. Be sure to include a docstring in your function.

```python
temp = np.arange(10,40,1)
res = array((3.239,3.118,3.004,2.897,2.795,2.700,2.610,2.526,2.446,2.371,2.300,2.233,2.169,2.110,2.053,2.000,1.950,1.902,1.857,1.815,1.774,1.736,1.700,1.666,1.634,1.603,1.574,1.547,1.521,1.496))
res1 = res[::-1]
temp1 = temp[::-1]
```

```python
def Temp(resistance):
    """This function will set a new function tempInterp for the 1d interpolation
    of res1,temp1, and then return the value of that function at the resistance
    that is given as the function argument."""
    tempInterp = interp1d(res1, temp1)
    temp2 = tempInterp(resistance)
    return temp2
```

```python
print Temp(1.75)
```

This is a value I would expect for the temperature at a resistance of 2.1

```python
Resistance = np.arange(1.5,3.239,0.01)

plt.plot(Resistance,Temp(Resistance), "r", label = "Temperature")
plt.legend(loc="upper right")
plt.title("Temperature vs Resistance", size = 15)
plt.xlabel("Resistance (mega-Ohms)", size = 13)
plt.ylabel("Temperature (degrees-Celcius)", size = 12)
plt.xlim(1,4)
plt.ylim(0,45)
t = 1.75
plt.annotate(r'Temperature(1.75 MOhms) = 30.6315',
            xy=(t, Temp(t)), xycoords='data',
            xytext=(+10, +20), textcoords='offset points', fontsize=12,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.show()
```
