# Dropper Simulation

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from IPython.html.widgets import interact, interactive
from IPython.html import widgets
from IPython.display import display
from IPython.display import clear_output
```

## Solving for a single droplet

### Derivative Function

Here is the derivative function that takes in an array, y, and then outputs the derivative of that array through a defined function for walker/bouncer motion. This is just the format that the Fortran based "odeint" solver uses. The functions used are:

\begin{equation}
\ddot{\mathbf{r}}_x = F_b\sin\left( \frac{2\pi\sqrt{\dot{\mathbf{x}}^2+\dot{\mathbf{y}}^2}}{v_f}\right) \frac{\dot{\mathbf{x}}}{\sqrt{\dot{\mathbf{x}}^2+\dot{\mathbf{y}}^2}}-f^v\dot{\mathbf{x}}
\end{equation}

\begin{equation}
\ddot{\mathbf{r}}_y = F_b\sin\left( \frac{2\pi\sqrt{\dot{\mathbf{x}}^2+\dot{\mathbf{y}}^2}}{v_f}\right) \frac{\dot{\mathbf{y}}}{\sqrt{\dot{\mathbf{x}}^2+\dot{\mathbf{y}}^2}}-f^v\dot{\mathbf{y}}
\end{equation}

```python
def derivdrop(y,t,Fb):
    vRx = y[0]
    vRy = y[2]
    Rx = y[1]
    Ry = y[3]
    
    vRmag = sqrt(vRx**2+vRy**2)
    Fbcrit = 3.2e-8  # Newtons
    m = 5e-7         # kg
    vFara = 0.19     # m/s
    dragC = 1e-6     # N*s/m
    
    dvRx_dt = (Fb*sin(2*pi*vRmag/vFara)*vRx/vRmag - dragC*vRx) / m
    dvRy_dt = (Fb*sin(2*pi*vRmag/vFara)*vRy/vRmag - dragC*vRy) / m
    
    return np.array([dvRx_dt, vRx, dvRy_dt, vRy])
```

### Plotting Function

Here is the plotting function that allows interact, it solves the ordinary differential equation given above and outputs the position and the velocity, which is then plotted as a position. This data is also used to graph the velocity and position as a function of time. 

```python
def plot_drop(init_vx = 0.035, init_vy = 0.035, init_x = 0.,
              init_y = 0., Fb = 3.2,t = 0.):
    vRx_0 = init_vx
    Rx0 = init_x
    vRy_0 = init_vy
    Ry0 = init_y
    initarray = np.array([vRx_0,Rx0,vRy_0,Ry0])
    Fb_val = Fb*10**-8
    
    ts = np.arange(0.,t+0.1,0.01)
    output = odeint(derivdrop, initarray, ts, args = (Fb_val,))
    
    vRx = output[:,0]
    vRy = output[:,2]
    Rx = output[:,1]
    Ry = output[:,3]
    
    mylabel =  r'$Dropper\ Simulation:\ Time\ =\ %.1f\ seconds$' % t
    print "Fb, the Forcing coefficient, is multiplied by 1e-8"
    figure(1)
    plt.subplot(2, 1, 1)
    plt.plot(ts,vRx, 'y.-')
    plt.title(r'$Velocity\ vectors$',size = 17.5)
    plt.ylabel('X-Component')

    plt.subplot(2, 1, 2)
    plt.plot(ts,vRy, 'r.-')
    plt.xlabel('time (s)')
    plt.ylabel('Y-Component')
    plt.show()
    
    plt.plot(Rx,Ry,'b--',label = "Walker/Bouncer")     
    plt.plot(Rx0,Ry0,'bo')                                
    plt.plot(Rx[-1],Ry[-1],'bd')
    plt.title(mylabel,size=17.5)                       
    plt.ylabel("Y",size=13)
    plt.xlabel("X",size=13)
    plt.xlim(-2,2)
    plt.ylim(-2,2)
    plt.grid()
    plt.legend(loc="best")
    plt.show()
```

```python
p = interact(plot_drop,init_vx = (-.1,.1,0.005), init_vy = (-.1,.1,0.005),
             init_x = (-2.,2.,0.1), init_y = (-2.,2.,0.1),
             t = (0.,100.,.1), Fb = (2,10,0.1))
```

# Solving for multiple droplets

We must now look at a new equation that incorporates the action at a distance given by the other droplets wave. We also look at a new derivative function that will now be an length-8 array that includes the two particles and the updated differential equation.

### Derivative Function

We will make only the forcing constant changable here. Another important aspect to be noted is that in the paper by Couder et al., the equation given has 3 different forcing constants. For sake of simplicity we will assume they are all equal.

```python
def deriv2drop(y,t,Fb):
    vx1 = y[0]
    vy1 = y[2]
    x1 = y[1]
    y1 = y[3]
    vx2 = y[4]
    vy2 = y[6]
    x2 = y[5]
    y2 = y[7]
    
    v1mag = sqrt(vx1**2+vy1**2)
    v2mag = sqrt(vx2**2+vy2**2)
    rhox1 = x1 - x2
    rhoy1 = y1 - y2
    rhox2 = x2 - x1
    rhoy2 = y2 - y1
    rho1sq = rhox1**2 + rhoy1**2
    rho2sq = rhox2**2 + rhoy2**2
    
    Fbcrit = 3.2e-8  # Newtons
    m = 5e-7         # kg
    vFara = 0.19     # m/s
    lambF = 4.53e-3  # m
    dragC = 1e-6     # N*s/m
    alpha = pi
    
    # Particle 1
    dvx1_dt = (Fb*sin(2*pi*v1mag/vFara)*vx1/v1mag 
               + alpha*Fb*rhox1/rho1sq*sin(2*pi/lambF*sqrt(rho1sq))
               - dragC*vx1) / m
    dvy1_dt = (Fb*sin(2*pi*v1mag/vFara)*vy1/v1mag 
               + alpha*Fb*rhoy1/rho1sq*sin(2*pi/lambF*sqrt(rho1sq))
               - dragC*vy1) / m
    
    # Particle 2
    dvx2_dt = (Fb*sin(2*pi*v2mag/vFara)*vx2/v2mag 
               + alpha*Fb*rhox2/rho2sq*sin(2*pi/lambF*sqrt(rho2sq))
               - dragC*vx2) / m
    dvy2_dt = (Fb*sin(2*pi*v2mag/vFara)*vy2/v2mag 
               + alpha*Fb*rhoy2/rho2sq*sin(2*pi/lambF*sqrt(rho2sq))
               - dragC*vy2) / m
    
    return np.array([dvx1_dt, vx1, dvy1_dt, vy1, dvx2_dt, vx2, dvy2_dt, vy2])
```

## Plotting 2-Drop function

```python
def plot_2drop(init_vx1 = 0.035, init_vy1 = 0., init_vx2 = -.035,
              init_vy2 = 0., Fb = 3.4, t = 0.):
    vx1_0 = init_vx1
    x1_0 = -0.5
    vy1_0 = init_vy1
    y1_0 = -0.5
    vx2_0 = init_vx2
    x2_0 = 0.5
    vy2_0 = init_vy2
    y2_0 = 0.5
    initarray = np.array([vx1_0,x1_0,vy1_0,y1_0,vx2_0,x2_0,vy2_0,y2_0])
    Fb_val = Fb*10**-8
    
    ts = np.arange(0.,t+0.1,0.01)
    output = odeint(deriv2drop, initarray, ts, args = (Fb_val,))
    
    vx1 = output[:,0]
    vy1 = output[:,2]
    x1 = output[:,1]
    y1 = output[:,3]
    vx2 = output[:,4]
    vy2 = output[:,6]
    x2 = output[:,5]
    y2 = output[:,7]
    
    mylabel =  r'$Dropper\ Simulation:\ Time\ =\ %.1f\ seconds$' % t
    mylabel2 =  "Velocity Graph: t = %.1f" % t
    velotit = r'$Velocity\ Graph:\ t = %.1f$' % t
    print "Fb, the Forcing coefficient, is multiplied by 1e-8"
        
    def on_button_clicked(b):
        plt.subplot(2, 1, 1)
        plt.plot(ts,vx1, 'r.-', label = "x-comp.")
        plt.plot(ts,vy1, 'g.-', label = "y-comp.")
        plt.title(velotit,size = 17.5)
        plt.legend(loc = "best")
        plt.ylabel('Drop 1 Velocity')

        plt.subplot(2, 1, 2)
        plt.plot(ts,vx2, 'y.-', label = "x-comp.")
        plt.plot(ts,vy2, 'b.-', label = "y-comp.")
        plt.ylabel('Drop 2 Velocity')
        plt.xlabel('time (s)')
        plt.legend(loc = "best")
        plt.show()
    
    button = widgets.ButtonWidget(description = mylabel2)
    display(button)
    button.on_click(on_button_clicked)
    
    plt.plot(x1,y1,'c--',label = "Droplet 1")
    plt.plot(x2,y2,'k--',label = "Droplet 2")
    plt.plot(x1[-1],y1[-1],'cd')                       
    plt.plot(x2[-1],y2[-1],'kd')
    plt.plot(x1[0],y1[0],'co')                       
    plt.plot(x2[0],y2[0],'ko')
    plt.title(mylabel,size=17.5)                       
    plt.ylabel("Y",size=13)
    plt.xlabel("X",size=13)
    plt.xlim(-2,2)
    plt.ylim(-2,2)
    plt.grid()
    plt.legend(loc="best")
    plt.show()
```

```python
p = interact(plot_2drop, init_vx1 = (-.1,.1,0.005), init_vy1 = (-.1,.1,0.005),
             init_vx2 = (-.1,.1,0.005), init_vy2 = (-.1,.1,0.005),
             t = (0.,100.,.1), Fb = (2,10,0.1))
```
