## Question 1

For this question refer to the solutions for the simple pendulum we got using scipy's odeint from the beginning of the ODE Tour.

(a) See what happens to the period of the pendulum as you change the initial angle of release. Plot θ vs t for 5 different initial angles, θ0 , spanning 0 to π. (i.e. 0, π/4 , π/2 , 3π/4 , and π ). Put them all on the same plot. What happens to the period of the pendulum as the initial angle increases from π/4 to 3π/4? What happens to the trajectory when the initial angle is 0 and π ?

```python
def deriv(theta_current,t):  # return derivatives of the array theta
    g = 9.8
    ell = 0.5
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/ell)*np.sin(theta)
    return np.array([dtheta, domega])

g = 9.8
ell = 0.5
t = np.arange(0.,10.+0.01,0.01)

omega0 = 0.  #Starting from rest, omega = 0
th0a = 0.
th0b = pi/4.
th0c = pi/2.
th0d = 3.*pi/4
th0e = pi

def theta_init(th0):
    theta_init = np.array([th0, omega0])
    theta = odeint(deriv,theta_init,t)
    return theta

a = theta_init(th0a)
b = theta_init(th0b)
c = theta_init(th0c)
d = theta_init(th0d)
e = theta_init(th0e)

plt.plot(t,a[:,0], label = r'$\theta_0=0$')
plt.plot(t,b[:,0], label = r'$\theta_0=\pi/4$')
plt.plot(t,c[:,0], label = r'$\theta_0=\pi/2$')
plt.plot(t,d[:,0], label = r'$\theta_0=3\pi/4$')
plt.plot(t,e[:,0], label = r'$\theta_0=\pi$')
plt.xlabel(r'$t\ (seconds)$',fontsize = 20)
plt.ylabel(r'$\theta\ (rad)$',fontsize = 20)
plt.ylim(-3,7)
plt.title(r'$Varying\ \theta_0$', size = 20)
plt.grid()
plt.legend()
plt.show()
```

(b) Make plots of θ vs. t, ω vs. t, and ω vs. θ (the so-called "phase plot") for θ0 = 0.95π. Look carefully at their behavior. What is going on in these figures?

```python
th0 = 0.95*pi
a = theta_init(th0)

plt.plot(t,a[:,0],label = r'$\theta$')
plt.title(r'$\theta\ vs.\ t$', size = 20)
plt.xlabel(r'$t$', size = 15)
plt.ylabel(r'$\theta$', size = 15)
plt.ylim(-4,5)
plt.legend()
plt.show()

plt.plot(t,a[:,1],'r',label = r'$\omega$')
plt.title(r'$\omega\ vs.\ t$', size = 20)
plt.xlabel(r'$t$', size = 15)
plt.ylabel(r'$\omega$', size = 15)
plt.ylim(-10,13)
plt.legend()
plt.show()

plt.plot(a[:,0],a[:,1],'g')
plt.title(r'$\omega\ vs.\ \theta$', size = 20)
plt.xlabel(r'$\omega$', size = 15)
plt.ylabel(r'$\theta$', size = 15)
plt.ylim(-10,13)
plt.xlim(-4,4)
plt.legend()
plt.show()
```

(c) For the θ0 = 0.95π case, check that energy is conserved by plotting E/m vs. t (where E/m=gℓcos(θ)+1/2ℓ^2ω^2 ).
Is the energy a constant? If not, then the computational errors are producing a non-physical loss (or gain) of energy. Review the documentation for odeint and change the error tolerance value to a tighter number.

```python
g = 9.8
ell = 0.5
th0 = 0.95*pi
t = np.arange(0.,10.+0.01,0.01)

b = theta_init(th0)
E = 9.81*0.5*(1-np.cos(b[:,0]))+(0.5)*(0.5*b[:,1])**2

plt.plot(t,E,'b',label = r'$E/m$')
plt.title(r'$E/m\ vs\ t$', size = 20)
plt.xlabel(r'$t$', size = 15)
plt.ylabel(r'$E/m$', size = 15)
plt.ylim(9,10)
plt.legend()
plt.show()
```

## Question 2

(a) Use Euler's method (not odeint) to solve the following first-order differential equation with given initial condition y(0) = 0.

\begin{equation}
\frac{dy}{dx} = x + 2y
\end{equation}

Find the value of y when x = 1, y(1), using two different step sizes: 0.25 then 0.02.

```python
def EulerODE(stepsize,value):
    xmax = 5 # maximum time
    dx = stepsize # time step
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
    dy = np.zeros(nsteps)
    # set initial conditions, redundant in this example
    x[0] = 0.
    y[0] = 0.
    dy[0] = x[0]+2*y[0]
    index = value/dx
    
    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        dy[i+1] = x[i] + 2*y[i] 
        y[i+1] = y[i] + dx*dy[i]

    return x,y,y[index]

x1,y1,y1a = EulerODE(0.25,1)
x2,y2,y2a = EulerODE(0.02,1)

print "y(1) for step size 0.25 =",y1a
print "y(1) for step size 0.02 =",y2a
```

(b) Plot the results from part (a) for the two different step sizes with the exact solution

\begin{equation}
y = 0.25 e^{2x} - 0.5 x - 0.25
\end{equation}

on one plot. Include a legend. Which step size gives a more accurate result?

```python
actual = lambda x: 0.25*exp(2*x)-0.5*x-0.25
x3 = np.arange(0,5.01,0.01)
y3 = actual(x3)
plt.plot(x1,y1,label = "Stepsize of 0.25")
plt.plot(x2,y2,label = "Stepsize of 0.02")
plt.plot(x3,y3,label = "Exact Solution")
plt.xlabel("x", size = 15)
plt.ylabel("y", size = 15)
plt.title("Graphing ODE's")
plt.xlim(0,2)
plt.ylim(0,3)
plt.legend(loc = "upper left")
plt.show()
```

The 0.02 stepsize gives a more accurate result. Which is something that you would expect.

(c) Now solve the same problem using Euler-Cromer and the Midpoint method for the same pair of step sizes and plot the results along with the exact solution. Then also plot the difference between the approximate numerical solutions and the exact solution. Which combination of method and step-size gives the best result?

```python
def EulerCromerODE(stepsize,value):
    xmax = 5 # maximum time
    dx = stepsize # time step
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
    dy = np.zeros(nsteps)
    # set initial conditions, redundant in this example
    x[0] = 0.
    y[0] = 0.
    dy[0] = x[0]+2*y[0]
    index = value/dx
    
    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        dy[i+1] = x[i] + 2*y[i] 
        y[i+1] = y[i] + dy[i+1]*dx

    return x,y,y[index]

x3,y3,y3a = EulerCromerODE(0.25,1)
x4,y4,y4a = EulerCromerODE(0.02,1)

print "Euler-Cromer Method"
print "y(1) for step size 0.25 =",y3a
print "y(1) for step size 0.02 =",y4a
print

def MidpointMethod(stepsize,value):
    xmax = 5 # maximum time
    dx = stepsize # time step
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
    dy = np.zeros(nsteps)
    # set initial conditions, redundant in this example
    x[0] = 0.
    y[0] = 0.
    dy[0] = x[0]+2*y[0]
    index = value/dx
    
    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        dy[i+1] = x[i] + 2*y[i] 
        y[i+1] = y[i] + 0.5*(dy[i]+dy[i+1])*dx

    return x,y,y[index]

x5,y5,y5a = MidpointMethod(0.25,1)
x6,y6,y6a = MidpointMethod(0.02,1)

print "Midpoint Method"
print "y(1) for step size 0.25 =",y5a
print "y(1) for step size 0.02 =",y6a
```

```python
actual = lambda x: 0.25*exp(2*x)-0.5*x-0.25
x7 = np.arange(0,5.01,0.01)
y7 = actual(x7)
## We can keep x1 and x2 in the problem becuase they just use the same
## step every time. We must redefine y's though.

plt.plot(x1,y1,label = "Euler: 0.25")
plt.plot(x2,y2,label = "Euler: 0.02")
plt.plot(x3,y3,label = "EulerCromer: 0.25")
plt.plot(x4,y4,label = "EulerCromer: 0.02")
plt.plot(x5,y5,label = "Midpoint: 0.25")
plt.plot(x6,y6,label = "Midpoint: 0.02")
plt.plot(x7,y7,label = "Exact Solution")
plt.xlabel("x", size = 15)
plt.ylabel("y", size = 15)
plt.title("Graphing ODE's")
plt.xlim(0,2)
plt.ylim(0,3)
plt.legend(loc = "upper left")
plt.show()
```

```python
ya= actual(x1)-y1 
yb= actual(x2)-y2
yc= actual(x3)-y3
yd= actual(x4)-y4 
ye= actual(x5)-y5
yf= actual(x6)-y6
```

```python
plt.plot(x1,ya,label = "Euler: 0.25")
plt.plot(x2,yb,label = "Euler: 0.02")
plt.plot(x3,yc,label = "EulerCromer: 0.25")
plt.plot(x4,yd,label = "EulerCromer: 0.02")
plt.plot(x5,ye,label = "Midpoint: 0.25")
plt.plot(x6,yf,label = "Midpoint: 0.02")
plt.xlabel(r'$x$', size = 15)
plt.ylabel(r'$\Delta\ y$', size = 15)
plt.xlim(0,2)
plt.ylim(0,10)
plt.title("Difference from Exact Solution")
plt.legend(loc = "upper left")
plt.show()
```

The Euler-Cromer method with a step of 0.02 gives the best result.

## Question 3

(a) Create a derivative function to solve this system of equations and use odeint to solve it for a projectile fired at an angle of 50 degrees from the horizontal at an initial speed of 30 m/s. Track the projectile's motion over the time interval of 0 to 200 seconds and let (x0,y0) = (0,0). You'll have to calculate (vx0,vy0) from the information provided. Let vter=44 m/s. Plot the trajectory of the projectile (y vs. x) to see how far it can travel in the x direction before landing.

```python
def derivQuadD(y,t,g,vter):
    vx = y[0]
    vy = y[2]
        
    x1 = y[1]
    y1 = y[3]
    dv_xdt = -g * (vx/(vter**2)) * sqrt(vx**2+vy**2)
    dv_ydt = -g * (1 + (vy/vter**2) * sqrt(vx**2+vy**2))
    
    return np.array([dv_xdt, vx, dv_ydt, vy])
```

```python
from scipy.interpolate import interp1d
from scipy.optimize import fsolve

def plot_quaddrag(init_v = 30.,init_angle = 50.,init_x = 0.,init_y = 0.,t = 200.,g = 9.8,vter = 44.):
    vx_0 = init_v*np.cos(init_angle/180. * np.pi)
    vy_0 = init_v*np.sin(init_angle/180. * np.pi)
    x0 = init_x
    y0 = init_y
    t = np.arange(0.,t+0.1,0.1)
    
    output = odeint(derivQuadD, np.array([vx_0,x0,vy_0,y0]), t, args=(g, vter))
    vx = output[:,0]
    vy = output[:,2]
    x = output[:,1]
    y = output[:,3]
    
    xnd = x0 + vx_0*t
    ynd = y0 +vy_0*t-1/2.*g*t**2
    
    plt.plot(x,y,label = "With Drag")
    plt.plot(xnd,ynd,label = "No Drag")
    plt.title("Quadratic Drag vs No Drag Trajectory")
    plt.ylabel("Height (m)")
    plt.xlabel("Distance (m)")
    plt.xlim(0,100)
    plt.ylim(0,40)
    plt.grid()
    plt.legend()
    plt.show()
    
    quadinterp = interp1d(x,y)
    nodraginterp = interp1d(xnd,ynd)
    zero = fsolve(quadinterp,100)
    zero2 = fsolve(nodraginterp,100)
    print "y(x)=0 for drag:",zero
    print "y(x)=0 for no drag:",zero2

plot_quaddrag(30.,50.,0.,0.,200.,9.81,44.)
```

```python
from IPython.html.widgets import interact, interactive

d = interact(plot_quaddrag,init_v = (0.,40.),init_angle = (0.,89.),
             init_y = (0.,20.),t = (0.,50.),vter = (0.,110.))
```
