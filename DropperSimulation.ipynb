# Dropper Simulation

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from IPython.html.widgets import interact, interactive
```

## Derivative Function

Here is the derivative function that takes in an array, y, and then outputs the derivative of that array through a defined function for walker/bouncer motion. This is just the format that the Fortran based "odeint" solver uses. The functions used are:

\begin{equation}
\ddot{\mathbf{r}}_x = F_b\sin\left( \frac{2\pi\sqrt{\dot{\mathbf{x}}^2+\dot{\mathbf{y}}^2}}{v_f}\right) \frac{\dot{\mathbf{x}}}{\sqrt{\dot{\mathbf{x}}^2+\dot{\mathbf{y}}^2}}-f^v\dot{\mathbf{x}}
\end{equation}

\begin{equation}
\ddot{\mathbf{r}}_y = F_b\sin\left( \frac{2\pi\sqrt{\dot{\mathbf{x}}^2+\dot{\mathbf{y}}^2}}{v_f}\right) \frac{\dot{\mathbf{y}}}{\sqrt{\dot{\mathbf{x}}^2+\dot{\mathbf{y}}^2}}-f^v\dot{\mathbf{y}}
\end{equation}

```python
def derivdrop(y,t,Fb):
    vRx = y[0]
    vRy = y[2]
    Rx = y[1]
    Ry = y[3]
    
    vRmag = sqrt(vRx**2+vRy**2)
    Fbcrit = 3.2e-8  # Newtons
    m = 5e-7         # kg
    vFara = 0.19     # m/s
    dragC = 1e-6     # N*s/m
    
    dvRx_dt = (Fb*sin(2*pi*vRmag/vFara)*vRx/vRmag - dragC*vRx) / m
    dvRy_dt = (Fb*sin(2*pi*vRmag/vFara)*vRy/vRmag - dragC*vRy) / m
    
    return np.array([dvRx_dt, vRx, dvRy_dt, vRy])
```

## Plotting Function

Here is the plotting function that allows interact, it solves the ordinary differential equation given above and outputs the position and the velocity, which is then plotted as a position. This data is also used to graph the velocity and position as a function of time. 

```python
def plot_drop(init_vx = 0.035, init_vy = 0.035, init_x = 0.,
              init_y = 0., Fb = 3.2,t = 0.):
    vRx_0 = init_vx
    Rx0 = init_x
    vRy_0 = init_vy
    Ry0 = init_y
    initarray = np.array([vRx_0,Rx0,vRy_0,Ry0])
    Fb_val = Fb*10**-8
    
    ts = np.arange(0.,t+0.1,0.01)
    output = odeint(derivdrop, initarray, ts, args = (Fb_val,))
    
    vRx = output[:,0]
    vRy = output[:,2]
    Rx = output[:,1]
    Ry = output[:,3]
    
    mylabel =  r'$Dropper\ Simulation:\ Time\ =\ %.1f\ seconds$' % t
    print "Fb, the Forcing coefficient, is multiplied by 1e-8"
    figure(1)
    plt.subplot(2, 1, 1)
    plt.plot(ts,vRx, 'y.-')
    plt.title(r'$Velocity\ vectors$',size = 17.5)
    plt.ylabel('X-Component')

    plt.subplot(2, 1, 2)
    plt.plot(ts,vRy, 'r.-')
    plt.xlabel('time (s)')
    plt.ylabel('Y-Component')
    plt.show()
    
    plt.plot(Rx,Ry,'b--',label = "Walker/Bouncer")     
    plt.plot(Rx0,Ry0,'bo')                                
    plt.plot(Rx[-1],Ry[-1],'bd')
    plt.title(mylabel,size=17.5)                       
    plt.ylabel("Y",size=13)
    plt.xlabel("X",size=13)
    plt.xlim(-2,2)
    plt.ylim(-2,2)
    plt.grid()
    plt.legend(loc="best")
    plt.show()
```

```python
p = interact(plot_drop,init_vx = (-.1,.1,0.005), init_vy = (-.1,.1,0.005),
             init_x = (-2.,2.,0.1), init_y = (-2.,2.,0.1),
             t = (0.,100.,.1), Fb = (2,10,0.1))
```
